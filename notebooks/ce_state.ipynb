```python
import pandas as pd
```

```python
with open('../data/slurm_wrapper_ce5.log') as f:
    data5 = ([line.split() for line in f])
with open('../data/slurm_wrapper_ce6.log') as f:
    data6 = ([line.split() for line in f])
```

```python
ce5 = pd.DataFrame([[pd.to_datetime(x[0] + 'T' + x[1]), x[4], float(x[10]), int(x[13]), ''.join(x[16:])] 
                       for x in data5], columns=['BEGIN', 'USER', 'DURATION', 'RC', 'COMMAND'])
ce6 = pd.DataFrame([[pd.to_datetime(x[0] + 'T' + x[1]), x[4], float(x[10]), int(x[13]), ''.join(x[16:])] 
                       for x in data6], columns=['BEGIN', 'USER', 'DURATION', 'RC', 'COMMAND'])
```

```python
ce5['value_grp'] = (ce5.RC.diff(1) != 0).cumsum()
ce6['value_grp'] = (ce6.RC.diff(1) != 0).cumsum()
```

```python
codes5_grouped = pd.DataFrame({
            'BeginDate' : ce5.groupby('value_grp').BEGIN.first(), 
              'EndDate' : ce5.groupby('value_grp').BEGIN.last(),
              'Consecutive' : ce5.groupby('value_grp').size(), 
              'Returncode' : ce5.groupby('value_grp').RC.first()})\
            .sort_values('Consecutive', ascending=False)
codes6_grouped = pd.DataFrame({
            'BeginDate' : ce6.groupby('value_grp').BEGIN.first(), 
              'EndDate' : ce6.groupby('value_grp').BEGIN.last(),
              'Consecutive' : ce6.groupby('value_grp').size(), 
              'Returncode' : ce6.groupby('value_grp').RC.first()})\
            .sort_values('Consecutive', ascending=False)
```

```python
group5_1 = codes5_grouped[codes5_grouped['Returncode'] == 1]
group6_1 = codes6_grouped[codes6_grouped['Returncode'] == 1]
```

```python
group5_1 = group5_1.sort_values('BeginDate', ascending=False)
group6_1 = group6_1.sort_values('BeginDate', ascending=False)
```
