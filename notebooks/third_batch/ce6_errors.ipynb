```python
import pandas as pd
```

```python
with open('../data/slurm_wrapper_ce6.log') as f:
    data6 = ([line.split() for line in f])
```

```python
ce6 = pd.DataFrame([[x[0] + 'T' + x[1], x[4], float(x[10]), int(x[13]), ''.join(x[16:])] 
                       for x in data6], columns=['BEGIN', 'USER', 'DURATION', 'RC', 'COMMAND'])
```

```python
ce6['value_grp'] = (ce6.RC.diff(1) != 0).cumsum()
```

```python
ce6 = ce6.reset_index()
```

```python
f_index = pd.DataFrame({'Consecutive' : ce6.groupby('value_grp').size(),
                    'index' : ce6.groupby('value_grp').index.first(),
                    'Returncode' : ce6.groupby('value_grp').RC.first()})
```

```python
fe_index = f_index[f_index['Returncode'] == 1]
```

```python
fe_df = ce6.join(fe_index['index'].to_frame().set_index('index'), how='inner').sort_values('BEGIN', ascending=False)
```
